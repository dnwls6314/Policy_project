```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pprint
import plotly
from sklearn.cluster import DBSCAN
from sklearn.manifold import TSNE
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_excel("./data/통합전처리_210105_군집분석용.xls", header=0)
df
```

```python
feature=df.drop(['학생통계번호'], axis=1)
```

```python
feature
```

```python
minmaxScaler=MinMaxScaler()
minmaxScaler.fit(feature)
```

```python
print(minmaxScaler.transform(feature))
```

```python
colnames=feature.columns
```

```python
new_colnames=[i+'_mms' for i in colnames]
aggregated_df=pd.concat([feature, pd.DataFrame(minmaxScaler.fit_transform(feature[colnames]), columns=new_colnames)], axis=1)
```

```python
aggregated_df
```

```python
aggregated_df.to_csv("./data/모든변수사용_MinMax칼럼추가_통합전처리.csv", encoding='cp949')
```

```python
feature_2 = aggregated_df[new_colnames]
```

```python
feature_2
```

```python
tsne_model = TSNE(learning_rate=300)
transformed = tsne_model.fit_transform(feature_2)
```

```python
# # 클러스터 15개
# dbscan_model = DBSCAN(eps=4, min_samples=20)
# predict = dbscan_model.fit(transformed)
# y_pred = predict.labels_
```

```python
# # 클러스터 9개
# dbscan_model = DBSCAN(eps=6, min_samples=20)
# predict = dbscan_model.fit(transformed)
# y_pred = predict.labels_
```

```python
# 클러스터 4개
dbscan_model = DBSCAN(eps=8, min_samples=20)
predict = dbscan_model.fit(transformed)
y_pred = predict.labels_
```

```python
# # 클러스터 6개
# dbscan_model = DBSCAN(eps=7.5, min_samples=30)
# predict = dbscan_model.fit(transformed)
# y_pred = predict.labels_
```

```python
df_describe = pd.DataFrame(y_pred)
df_describe.describe()
```

```python
dataset = pd.DataFrame({'x_component':transformed[:,0],'y_component':transformed[:,1]})
dataset['cluster_num'] = pd.Series(predict.labels_)
```

```python
dataset
```

```python
final_df=pd.concat([df['학생통계번호'],feature],axis=1)
final_df=pd.concat([final_df,feature_2],axis=1)
final_df=pd.concat([final_df,dataset],axis=1)
final_df
```

```python
import plotly.express as px
    
fig = px.scatter(final_df, x="x_component", y="y_component", hover_name="학생통계번호", color = "cluster_num", size_max=60)
fig.update_layout(
     height=800)
fig.show()
```

```python
final_df.to_excel("./data/통합전처리_단과대제거_모든변수사용_클러스터4개.xlsx")
```
